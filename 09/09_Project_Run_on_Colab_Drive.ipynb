<details>
<summary>⚠️重要⚠️ `functions.py` と `analysis.py` の役割</summary>

今回のプロジェクトでは、プログラムを `functions.py` と `analysis.py` の2つのファイルに分けています。

```text
project/
├── functions.py
└── analysis.py
```

それぞれの役割は次の通りです。

| ファイル           | 役割                                      |
| -------------- | --------------------------------------- |
| `functions.py` | 解析に使う部品となる関数をまとめるファイル                   |
| `analysis.py`  | `functions.py` の関数を組み合わせて、解析全体を実行するファイル |

### `functions.py` の役割

`functions.py` には、解析の中で使う「部品」となる関数が書かれています。

たとえば、次のような処理です。

```text
都市名から緯度・経度を取り出す
Web APIから気温データを取得する
取得したデータをCSVファイルとして保存する
CSVファイルを読み込む
指定した月だけを取り出して、年ごとの平均気温を計算する
計算結果をグラフにする
```

それぞれの処理を関数として分けておくことで、同じ処理を何度も使い回しやすくなります。

たとえば、都市を変えたり、月を変えたりしても、データ取得・保存・計算・作図の流れは同じです。
そのため、共通して使う処理を `functions.py` にまとめています。

### `analysis.py` の役割

`analysis.py` は、`functions.py` に書かれた関数を読み込んで、解析全体の流れを組み立てるファイルです。

`analysis.py` の中では、次のような順番で処理が進みます。

```text
1. 解析したい都市と月を決める
2. 都市名から緯度・経度を取得する
3. Web APIから日平均気温データを取得する
4. データをCSVファイルとして保存する
5. CSVファイルを読み込む
6. 指定した月の年平均気温を計算する
7. グラフを作成して保存する
```

つまり、`functions.py` が「部品置き場」だとすると、`analysis.py` は「部品を順番に使って、1つの解析を完成させるファイル」です。

### なぜファイルを分けるのか

すべてのコードを1つのファイルや1つのセルに書くこともできます。
しかし、処理が長くなると、どこで何をしているのか分かりにくくなります。

そこで、次のように役割を分けます。

```text
functions.py
→ 何度も使う処理を関数としてまとめる

analysis.py
→ 関数を組み合わせて、解析全体の流れを書く
```

このように分けると、プログラムの見通しがよくなります。
また、あとから修正したり、別の条件で解析したりするときにも使いやすくなります。
</details>

<details>
<summary>補足: `if __name__ == "__main__":` の意味</summary>

`analysis.py` の最後には、次のようなコードがあります。

```python
if __name__ == "__main__":
    main()
```

これは、`analysis.py` が「直接実行されたときだけ」`main()` を実行するための書き方です。

### 直接実行する場合

たとえば、次のように `analysis.py` をPythonプログラムとして実行したとします。

```python
!python informatics-2026/09/project/analysis.py
```

この場合、Pythonは `analysis.py` を「直接実行されたファイル」として扱います。

そのとき、Pythonの中では次の条件が成り立ちます。

```python
__name__ == "__main__"
```

そのため、次の部分が実行されます。

```python
main()
```

つまり、`analysis.py` を直接実行すると、`main()` 関数が動き、解析が始まります。

### 他のファイルから読み込む場合

一方で、ノートブックや別のPythonファイルから、`analysis.py` の中の関数だけを使いたい場合もあります。

たとえば、次のように書く場合です。

```python
from analysis import run_temperature_analysis
```

この場合は、`analysis.py` を直接実行しているのではなく、`analysis.py` の中にある関数を読み込んでいます。

このときは、

```python
__name__ == "__main__"
```

にはなりません。

そのため、`main()` は自動では実行されません。

### なぜこの書き方を使うのか

この書き方を使うと、同じ `analysis.py` を2通りの使い方で利用できます。

```text
1. analysis.py を直接実行する
   → main() が実行され、決められた条件で解析が始まる

2. analysis.py から関数だけを読み込む
   → main() は自動実行されず、必要な関数だけを使える
```

今回の授業では、まず次のように直接実行します。

```python
!python informatics-2026/09/project/analysis.py
```

その後、条件を変えたい場合には、ノートブックから関数を読み込んで実行できます。

```python
from analysis import run_temperature_analysis

df_result = run_temperature_analysis("Sapporo", 1)
df_result.head()
```

このように、`if __name__ == "__main__":` を使うことで、
「ファイルとして実行する場合」と「関数を読み込んで使う場合」を分けることができます。
</details>

## 6. `__pycache__` について

Pythonファイルを実行したあと、プロジェクトフォルダの中に `__pycache__` というフォルダができることがあります。

これはエラーではありません。

`__pycache__` は、Pythonがプログラムを少し速く読み込むために自動で作る補助的なフォルダです。  
自分で編集する必要はありません。

今回の授業では、`__pycache__` は無視して構いません。


<details>
<summary>補足：なぜ __pycache__ ができるのか？</summary>

Pythonでは、別のPythonファイルから関数などを読み込むことがあります。

今回のプロジェクトでは、`analysis.py` の中で `functions.py` の関数を読み込んで使っています。

このようなとき、Pythonは読み込んだファイルを次回以降に使いやすい形で一時保存することがあります。  
その保存場所が `__pycache__` フォルダです。

たとえば、次のようなフォルダができることがあります。

```text
informatics-2026/09/project/__pycache__/
```

これはPythonが自動で作るものなので、学生が手で作る必要はありません。  
また、消してしまっても、必要になればPythonがまた自動で作ります。

</details>
