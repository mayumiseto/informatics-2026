## 2. 教材フォルダを用意する

`informatics-2026` フォルダをGoogle Driveに用意します。

前回までに作成した `MyDrive/informatics-2026` フォルダは削除しません。  
今回は、GitHub の最新版教材から `10` フォルダだけを取り出して、`MyDrive/informatics-2026/10` にコピーします。


````python
%%bash
set -e

cd /content/drive/MyDrive

# 既存の informatics-2026 フォルダは削除しない
if [ ! -d informatics-2026 ]; then
    mkdir informatics-2026
    echo "informatics-2026 フォルダを作成しました。"
else
    echo "informatics-2026 フォルダはすでにあります。削除せずにそのまま使います。"
fi

# GitHub の最新版を /content に一時ダウンロード
cd /content
rm -rf informatics-2026-main
rm -f informatics-2026-main.zip

wget -O informatics-2026-main.zip https://github.com/mayumiseto/informatics-2026/archive/refs/heads/main.zip
unzip -q informatics-2026-main.zip

if [ ! -d /content/informatics-2026-main/10 ]; then
    echo "ERROR: GitHubから取得した教材の中に 10 フォルダがありません。"
    echo "ローカルの 10 フォルダを git add / commit / push していない可能性があります。"
    echo "取得できたフォルダ:"
    ls -la /content/informatics-2026-main
    exit 1
fi

# 10 フォルダだけ MyDrive/informatics-2026 にコピー
cd /content/drive/MyDrive/informatics-2026

if [ -d 10 ]; then
    echo "既存の 10 フォルダを削除して、新しい 10 フォルダに置き換えます。"
    rm -rf 10
fi

cp -r /content/informatics-2026-main/10 ./10

echo ""
echo "第10回フォルダを準備しました。"
echo "作成場所: /content/drive/MyDrive/informatics-2026/10"
echo "09 フォルダは削除・変更していません。"
echo ""
echo "10 フォルダの中身:"
ls -la /content/drive/MyDrive/informatics-2026/10

````

## 9. 提出前チェック

必要なファイルと出力ファイルを確認します。


````python
%cd /content/drive/MyDrive/informatics-2026

from pathlib import Path

required_files = [
    "10/project/functions.py",
    "10/project/analysis.py",
]

print("必須ファイル:")
for path in required_files:
    if Path(path).exists():
        print("OK:", path)
    else:
        print("MISSING:", path)

csv_files = list(Path("10/project/output").glob("*.csv"))
png_files = list(Path("10/project/figures").glob("*.png"))

print("\nCSVファイル:")
if csv_files:
    for path in csv_files:
        print("OK:", path)
else:
    print("MISSING: CSVファイルが見つかりません")

print("\nPNGファイル:")
if png_files:
    for path in png_files:
        print("OK:", path)
else:
    print("MISSING: PNGファイルが見つかりません")

````
