## 2. `09_retry` フォルダをMyDriveに準備する

次のセルは、GitHubの最新版から `09_retry` フォルダだけを取り出して、次の場所にコピーします。

```text
/content/drive/MyDrive/informatics-2026/09_retry
```

既に `09_retry` がある場合は、上書きしません。


````python
%%bash
set -e

cd /content/drive/MyDrive

if [ ! -d informatics-2026 ]; then
    mkdir informatics-2026
    echo "informatics-2026 フォルダを作成しました。"
else
    echo "informatics-2026 フォルダはすでにあります。削除せずにそのまま使います。"
fi

cd /content
rm -rf informatics-2026-main
rm -f informatics-2026-main.zip

wget -O informatics-2026-main.zip https://github.com/mayumiseto/informatics-2026/archive/refs/heads/main.zip
unzip -q informatics-2026-main.zip

if [ ! -d /content/informatics-2026-main/09_retry ]; then
    echo "ERROR: GitHubから取得した教材の中に 09_retry フォルダがありません。"
    echo "教員側で 09_retry をGitHubにpushできているか確認してください。"
    echo "取得できたフォルダ:"
    ls -la /content/informatics-2026-main
    exit 1
fi

cd /content/drive/MyDrive/informatics-2026

if [ -d 09_retry ]; then
    echo "09_retry フォルダはすでにあります。"
    echo "上書きすると作業内容が消える可能性があるため、このセルでは上書きしません。"
else
    cp -r /content/informatics-2026-main/09_retry ./09_retry
    echo "09_retry フォルダを作成しました。"
fi

echo ""
echo "現在のフォルダ:"
ls -la /content/drive/MyDrive/informatics-2026
````
